```python
fare_survived=[]
fare_deaths=[]
data=[]
for lab,row in titanic_df.iterrows():
    if row["survived"]==1:
        fare_survived.append(row["fare"])
    else:
        fare_deaths.append(row["fare"])
fare_survived
fare_deaths
data.append(fare_survived)
data.append(fare_deaths)
fig=plt.figure(figsize=(10,7))
ax=fig.add_subplot(111)

bp_fare=ax.boxplot(data,patch_artist =True, notch="True", vert=0)

colors=["green","black"]

for patch,color in zip(bp_fare["boxes"], colors):
    patch.set_facecolor(color)
    
for whisker in bp_fare["whiskers"]:
    whisker.set(color = "blue",linewidth=1.5,linestyle=":")
    
for cap in bp_fare["caps"]:
    cap.set(color="blue", linewidth=2)
    
for median in bp_fare["medians"]:
    median.set( color="red", linewidth=3)
for flier in bp_fare["fliers"]:
    flier.set(marker="D", color="pink", alpha=0.5)
    
ax.set_yticklabels(["Survivers","Deaths"])
plt.title("Fare")
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.show()
```
